### Se importara lo necesario para hacer el ejercicio en este caso ocuparemos el csv, matplotlib y numpy
#### Con fechas se crea un arreglo donde se guardara la variable fechas
#### Con datos se creara un diccionario donde podre guardar las fechas  

```python
fechas=[]
datos={}
import csv
import matplotlib.pyplot as plt
import numpy as np
```

### Con los datos entregados se abrira la carpeta como archivo CSV

```python
with open('CasosGeneroEtario.csv') as casos:
    linea= csv.reader(casos, delimiter=',')
    variable=1
    contador=1
    for columna in linea:
        grupo_etario=columna.pop(0)
        Sexo=columna.pop(0)
        if variable==1:
            fechas=columna
            variable=0
        else:
            valores_enteros=list(map(int, columna))
            contador=0
            for fecha in fechas:
                try:
                    datos[fecha][Sexo]=valores_enteros[contador]
                    datos[fecha]['F']=valores_enteros[contador]
                except KeyError:
                    datos[fecha]={}
                    datos[fecha][Sexo]=valores_enteros[contador]  
                contador+=1
```

### Ahora con el matplotlib.pyplot se podra graficar los resultados echos recientemente
##### - con ax.set_title se podra colocar el titulo del grafico
##### - con  ax.set_xlabel se podra colocar el titulo para el eje x del grafico
##### - con ax.set_ylabel se podra colocar el titulo para el eje y del grafico


```python
fig, ax = plt.subplots(figsize=(100,10))
etiquetas=fechas
ejeX = np.arange(len(etiquetas))
rangos=list(datos['2020-03-25'].keys())
x={}
for rango in rangos:
    x[rango]=[]
for fecha in fechas:
    for rango in rangos:
        x[rango].append(datos[fecha][rango])
for rango in rangos:
    ax.plot(x[rango],label=rango)
ax.set_title('Infectados por Sexo en el tiempo') 
ax.set_xlabel('Fechas Contagio COVID')
ax.set_ylabel('Cantidad de Contagiados')
ax.set_xticks(ejeX)
ax.set_xticklabels(etiquetas,rotation=60)
ax.legend()
```

```python
fechas=[]
datos={}
with open('CasosGeneroEtario.csv') as casos:
    linea= csv.reader(casos, delimiter=',')
    variable=1
    contador=1
    for columna in linea:
        grupo_etario=columna.pop(0)
        sexo=columna.pop(0)
        if variable==1:
            fechas=columna
            variable=0
        else:
            valores_enteros=list(map(int, columna))
            contador=0
            for fecha in fechas:
                try:
                    datos[fecha][grupo_etario]=valores_enteros[contador]
                except KeyError:
                    datos[fecha]={}
                    datos[fecha][grupo_etario]=valores_enteros[contador]
                contador+=1   
```

##### Con ayuda del matplot se debera graficar en el eje x todas las fechas del grupo etario, para asi verificar como a crecido los contagios en el tiempo del grupo etario
