```python
covid_Brasil=casos_covid.query("Country_Region=='Brazil'")
covid_Brasil
```

```python
ggplot(covid_Brasil, aes(x="c", y="Confirmed", fill="c"))+geom_col()+theme(axis_text_x = element_text(angle = 45, vjust =1, hjust=1))+labs(x="Regiones",y="Total Confirmados", title="Covid")
```

```python
ggplot(covid_Brasil, aes(x="c", y="Deaths", fill="c"))+geom_col()+theme(axis_text_x = element_text(angle = 45, vjust =1, hjust=1))+labs(x="Regiones",y="Total Confirmados", title="Covid")
```

```python
chile=casos_covid.query("Country_Region=='Chile'")
chile
```

```python
ggplot(chile, aes(x="c", y="Confirmed", group=1))+geom_line()+geom_point()+theme(axis_text_x = element_text(angle = 45, vjust =1, hjust=1))+labs(x="Regiones", y="Total Confirmados",  title="COVID")
```

```python
ggplot(chile, aes(x="c", y="Deaths", group=1, fill="c"))+geom_line()+geom_point()+theme(axis_text_x = element_text(angle = 45, vjust =1, hjust=1))+labs(x="Regiones", y="Total Confirmados",  title="COVID")
```
